1. l'ammontare del reddito da lavoro dipendente (colonna L del dataset) per ogni regione e relativo grafico a torta

```python
#per ogni... sempre gropupby()
#reset index resetta l'index ad un intero 
es1 = redditi.groupby('Regione').sum()[['Reddito da lavoro dipendente e assimilati - Ammontare in euro']].reset_index().sort_values('Reddito da lavoro dipendente e assimilati - Ammontare in euro', ascending=False)
es1
```

```python
fig,ax = plt.subplots(figsize=(20,20))
ax.pie(es1["Reddito da lavoro dipendente e assimilati - Ammontare in euro"],labels=es1["Regione"],autopct= '%.2f%%')
plt.show()
```

2. l'ammontare del reddito da lavoro dipendente per ogni provincia e relativo grafico a barre verticali

```python
#merge unisce in questo caso provincie e redditi 
#
es2=pd.merge(province, redditi, how="right", left_on='SIGLA', right_on='Sigla Provincia')
es2
```

```python
es2Grafico= es2.groupby(["DEN_UTS"]).sum()
es2Grafico=es2Grafico[["Reddito da lavoro dipendente e assimilati - Ammontare in euro"]].reset_index()
es2Grafico
```

```python
fig,ax = plt.subplots(figsize=(30,8))
ax.bar(es2Grafico["DEN_UTS"],es2Grafico["Reddito da lavoro dipendente e assimilati - Ammontare in euro"])
plt.xticks(rotation=90)
plt.show()
```

6. dato il nome di una regione, inserito dall'utente, avere l'elenco delle province della regione con il reddito complessivo e il reddito medio

```python
cercare=input("inserire regione: ")
Regione_cercata=es2[es2["Regione"]==cercare]
a=Regione_cercata.groupby("Regione")['Reddito da lavoro dipendente e assimilati - Ammontare in euro'].sum()
b=Regione_cercata.groupby("Regione")['Reddito da lavoro dipendente e assimilati - Ammontare in euro'].mean()
print("Ammontare totale del reddito complessivo: ",a, " ammontare medio del reddito: ",b)
Regione_cercata["DEN_UTS"]
```

7. dato il nome di una provincia, inserito dall'utente, avere l'elenco dei comuni della provincia con il reddito. Ordinare in ordine crescente sul reddito. 

```python
cercare=input("inserire provincia: ")
Provincia_cercata=es2[es2["DEN_UTS"]==cercare]
a=Provincia_cercata.groupby("DEN_UTS")['Reddito da lavoro dipendente e assimilati - Ammontare in euro'].sum()
b=Provincia_cercata.groupby("DEN_UTS")['Reddito da lavoro dipendente e assimilati - Ammontare in euro'].mean()
print("Ammontare totale del reddito complessivo: ",a, " ammontare medio del reddito: ",b)
Provincia_cercata["Denominazione Comune"]
```

8. trovare la regione a cui appartiene il comune con il reddito più basso: visualizzare la mappa del comune e della regione

```python
Redditi_RegioniGeo=pd.merge(regioni, redditi, how="left", left_on='DEN_REG', right_on='Regione')
Redditi_RegioniGeo
```

```python
Redditi_ComGeo=pd.merge(comuni, redditi, how="left", left_on='COMUNE', right_on='Denominazione Comune')
Redditi_ComGeo
```

```python
Redditi_ProvGeo=pd.merge(province, redditi, how="left", left_on='SIGLA', right_on='Sigla Provincia')
Redditi_ProvGeo
```

```python
!pip install contextily
import contextily
```

```python
ComuneMin= Redditi_ProvGeo[Redditi_ProvGeo["Reddito da lavoro dipendente e assimilati - Ammontare in euro"]==Redditi_ProvGeo["Reddito da lavoro dipendente e assimilati - Ammontare in euro"].min()]
print(ComuneMin[["Regione","Denominazione Comune"]])
comUn=ComuneMin.unary_union
comReg=regioni[regioni.contains(comUn)]
ax = comReg.to_crs(epsg=3857).plot(figsize=(15,15), alpha= 0.4, edgecolor= "k", linewidth= 3)
ComuneMin.to_crs(epsg=3857).plot(ax=ax, alpha= 0.4, edgecolor= "k", linewidth= 3)
contextily.add_basemap(ax)
```
